<script>
  const resultElement = document.getElementById('result');

  function appendToResult(value) {
    resultElement.value += value;
  }

  function clearResult() {
    resultElement.value = '';
  }

  function calculate() {
    try {
      resultElement.value = eval(resultElement.value);
    } catch (error) {
      resultElement.value = 'Error';
    }
  }
</script>
